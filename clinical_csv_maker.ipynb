```python
import json
import pandas as pd
import glob
```

```python
with open('./menifest.txt','r') as f:
    columns = f.readline()
    column_list = columns.split('\t')
    c = column_list[-1].split('\n')[0]
    column_list[-1] = c
    menifest = {}
    for i in column_list:
        menifest[i] = []
    while True:
        line = f.readline()
        if not line:
            break
        else:
            line_list = line.split('\t')
            l = line_list[-1].split('\n')[0]
            line_list[-1] = l
            for c, l in zip(column_list, line_list):
                menifest[c].append(l)
                
df = pd.DataFrame(menifest)
df.to_csv('./menifest.csv', index = False)
```
